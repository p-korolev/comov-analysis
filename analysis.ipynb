# An Informal Analysis on Intra-sector Comovement

### Overview and Terminology
In this brief analysis, the aim is to pick up some sort of intuition on price divergence between same sector equities. Particularly, we will analyze the *comovement* between a sector leader $L$ and a sector follower $F$. In this case, $L$ and $F$ are simply two equities that belong to some market sector $M$. As seen in the **README** file, we can conceptualize a market sector $M$ as a collection of unique equities $E_1, E_2, ..., E_n$ where:

$$
\text{MarketCap}(E_i) \neq \text{MarketCap}(E_j) \text{ for all } i, j \leq n \text{ with } i \neq j
$$

In this case, $M \sub W$ where $W$ is an entire market. For example, Take the Energy sector of the American market. We can denote this sector as our sector $M \sub W:=\text{American Market}$. In a similar way, we can write an equity $E \in M$ if the respective company of $E$ serves the sector $M$. 

### Basic Results
From our assumption on the $\text{MarketCap}$ of assets $E_i \in M$, we can know that there exists some $E_l, E_f \in M$ such that:

$$
\text{(1) } E_l = \arg\max_{E \in M} \text{MarketCap}(E) \newline

\text{(2) } E_f = \arg\min_{E \in M} \text{MarketCap}(E)
$$

When trying to identify sector leaders, it is not generally the case that we choose the equity with the highest market cap in $M$. Typically, a sector will have numerous leaders $L_1, ..., L_k \in M$, where each of $L_i$'s market cap is greater than the average market cap for an equity in that sector. For the rest of the notebook, we write $w(e)$ or $w_e$ as the market cap of equity $e$. Then we have,

$$
w(L_m) > \frac{1}{n} \sum_{i=1}^{n} w(E_i)
$$

, for all $m \leq k$ and $E \in M$. These *leader* equities are like sector representatives; if the performance of the overall sector $M$ is unknown, we can look at the performance the $k$ largest companies belonging to $M$. Although not completely accurate, it would give us a good enough guess on the state of the sector. To formalize the intuition here, we know that the return (performance) of a sector $M:=\set{E_1,...,E_n}$ can be written in terms of a market cap weighted averages:

$$
R_M(t) = \frac{\sum_{i=1}^{n}w(E_i)R_{E_i}(t)}{\sum_{i=1}^{n}w(E_i)}
$$

Taking the difference between the general sector return and that of only the top $k$ companies ($k$ largest market caps), we get an error term $\epsilon$:

$$
\epsilon(t) = R_M(t) - \frac{\sum_{i=1}^{k}w(E_i)R_{E_i}(t)}{\sum_{i=1}^{k}w(E_i)} = \frac{\sum_{i=k+1}^{n}w(E_i)R_{E_i}(t)}{\sum_{i=1}^{n}w(E_i)}
$$

Since the sum of the market caps past the top $k$ equities is much smaller than the sum of market caps of all equities, $\epsilon$ is almost negligible. From here, we can assume that the returns of the top $k$ companies in sector $M$ represent the overall sector performance almost perfectly.

### Leader-Follower Comovement

For a market sector $M=\set{E_i}$, choose a representative $L$ from a list of potential sector representatives. Likewise, we can choose a sector follower $F \in M$, to be any other equity such that its market cap, $w_F$, is less than that of the average market cap of the sector, $\mu(w_M)$. 

From earlier, we know that any sector leader has a larger market cap than the sector average. One way to determine a clear sector leader is to choose $L$ such that: 

$$
L \in \set{E_l | w(E_i)>\mu(w_M)+\alpha \text{, } \alpha \in \mathbb{R_{>0}}} \sub M
$$

**Example**

Let $M=\set{E_1, E_2, E_3, E_4}$ with $w(E_1)=200$, $w(E_2)=170$, $w(E_3)=110$, $w(E_4)=98$. From here we calculate the average marketcap for an equity in $M$:

$$
\begin{aligned}
\mu(w_M) &= \frac{1}{n}\sum_{i=1}^{4} w(E_i) \\ 
&= \frac{1}{4}(200 + 170 + 110 + 98) \\
&= 144.5
\end{aligned}
$$

If we choose $\alpha=20$, then we can pick a clear sector leader to be $L = E_1$ or $L = E_2$.

Using the same set of rules to choose a sector follower $F$, we can define our follower to be some equity such that,

$$
F \in \set{E_f | w(E_f)<\mu(w_M)-\beta, \beta \in \mathbb{R_{>0}}} \sub M
$$
